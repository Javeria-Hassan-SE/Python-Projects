```python
import cv2

# Load the graph image
image = cv2.imread(r'C:\Users\HP\Downloads\graph3.png')

# Define the coordinates and labels for different elements
title_label = 'Graph Title'
x_label = 'X-Axis Label'
y_label = 'Y-Axis Label'

# Calculate the positions of the elements
title_coords = (image.shape[1] // 2 - len(title_label) * 6, 50)
x_label_coords = (image.shape[1] // 2 - len(x_label) * 6, image.shape[0] - 20)
y_label_coords = (20, image.shape[0] // 2)

# Draw rectangles and labels on the image
cv2.rectangle(image, (title_coords[0] - 10, title_coords[1] - 30), (title_coords[0] + len(title_label) * 12, title_coords[1] + 10), (0, 0, 255))
cv2.putText(image, title_label, title_coords, cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 0), 2)

cv2.putText(image, x_label, x_label_coords, cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 0), 2)
cv2.putText(image, y_label, y_label_coords, cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 0), 2)

# Display the labeled image
cv2.imshow('Labeled Graph', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

# Load the graph image
image = cv2.imread(r'C:\Users\HP\Downloads\graph3.png')

# Define the coordinates and labels for different elements
title_label = 'Graph Title'
x_label = 'X-Axis Label'
y_label = 'Y-Axis Label'

# Calculate the positions of the elements
title_coords = (image.shape[1] // 2 - len(title_label) * 6, 50)
x_label_coords = (image.shape[1] // 2 - len(x_label) * 6, image.shape[0] - 20)
y_label_coords = (20, image.shape[0] // 2)

# Draw rectangles and labels on the image
cv2.rectangle(image, (title_coords[0] - 10, title_coords[1] - 30), (title_coords[0] + len(title_label) * 12, title_coords[1] + 10), (0, 0, 255), cv2.FILLED)
cv2.putText(image, title_label, title_coords, cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)

cv2.putText(image, x_label, x_label_coords, cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 0), 2)

# Rotate and position the y-axis label
(y_label_height, y_label_width), _ = cv2.getTextSize(y_label, cv2.FONT_HERSHEY_SIMPLEX, 0.8, 2)
M = cv2.getRotationMatrix2D((y_label_coords[0], y_label_coords[1]), 90, 1)
rotated_label = cv2.warpAffine(cv2.putText(cv2.cvtColor(image, cv2.COLOR_BGR2RGB), y_label, y_label_coords, cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 0), 2), M, (y_label_height, y_label_width))

# Rotate the image back to its original orientation
rotated_image = cv2.rotate(rotated_label, cv2.ROTATE_90_COUNTERCLOCKWISE)

# Display the labeled image
cv2.imshow('Labeled Graph', rotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

# Load the graph image
image = cv2.imread(r'C:\Users\HP\Downloads\graph4.png')

# Define the coordinates and labels for different elements
title_label = 'Graph Title'
x_label = 'X-Axis Label'
y_label = 'Y-Axis Label'

# Calculate the positions of the elements
title_coords = (image.shape[1] // 2 - len(title_label) * 6, 50)
x_label_coords = (image.shape[1] // 2 - len(x_label) * 6, image.shape[0] - 20)
y_label_coords = (20, image.shape[0] // 2)

# Draw rectangles and labels on the image
cv2.rectangle(image, (title_coords[0] - 10, title_coords[1] - 30), (title_coords[0] + len(title_label) * 12, title_coords[1] + 10), (0, 0, 255))
cv2.putText(image, title_label, title_coords, cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 0), 2)

cv2.putText(image, x_label, x_label_coords, cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 0), 2)
cv2.putText(image, y_label, y_label_coords, cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 0), 2)

# Detect and label legends (if present)
legends = ['Legend 1']  # Replace with the actual legends or use a legend detection algorithm

legend_start_y = 100  # Initial y-coordinate for legends
legend_gap = 30  # Gap between consecutive legends

for legend_text in legends:
    legend_width = len(legend_text) * 8
    if image.shape[1] - legend_width > legend_width:
        # Legends on the right side
        legend_coords = (image.shape[1] - legend_width - 20, legend_start_y)
    else:
        # Legends on the left side
        legend_coords = (20, legend_start_y)
    cv2.putText(image, legend_text, legend_coords, cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 0), 2)
    legend_start_y += legend_gap

# Display the labeled image
cv2.imshow('Labeled Graph', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
