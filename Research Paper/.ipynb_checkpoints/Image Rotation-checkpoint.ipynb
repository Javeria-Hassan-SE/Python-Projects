```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
from skimage.transform import resize
from skimage.segmentation import slic
from skimage.segmentation import mark_boundaries

# Load the image
image_path = 'fox.jpg'  # Replace with the path to your image
image = io.imread(image_path)

# Resize the image for faster computation (optional)
resized_image = resize(image, (300, 300), anti_aliasing=True)

# Apply SLIC superpixel segmentation
segments = slic(resized_image, n_segments=100, compactness=10)

# Generate the superpixel map
superpixel_map = mark_boundaries(resized_image, segments)

# Calculate the saliency map
saliency_map = np.zeros_like(resized_image[:, :, 0], dtype=float)
for segment_id in np.unique(segments):
    mask = segments == segment_id
    saliency_map += np.mean(resized_image[:, :, 0][mask]) * mask

# Normalize the saliency map
saliency_map /= np.max(saliency_map)

# Display the original image and the saliency map
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
ax[0].imshow(resized_image)
ax[0].set_title('Original Image')
ax[0].axis('off')
ax[1].imshow(saliency_map)
ax[1].set_title('Saliency Map')
ax[1].axis('off')
plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np

def rotate_image(image, angle):
    # Get image shape and center
    height, width = image.shape[:2]
    center = (width / 2, height / 2)

    # Define the rotation matrix
    rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)

    # Perform the actual rotation
    rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))

    return rotated_image

# Load the image
image_path = 'path_to_image.jpg'  # Replace with the path to your image
image = cv2.imread(image_path)

# Define the desired rotation angle (in degrees)
angle = 45

# Rotate the image
rotated_image = rotate_image(image, angle)

# Display the original and rotated images
cv2.imshow('Original Image', image)
cv2.imshow('Rotated Image', rotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import numpy as np
from PIL import Image
from scipy.ndimage import rotate
import numpy as np

def rotate_image(image_path, angle, axis):
    # Load the image
    image = Image.open(image_path)

    # Convert the image to a numpy array
    image_array = np.array(image)

    # Determine the rotation axes
    if axis == 'x':
        axes = (1, 2)  # Rotate along the x-axis
    elif axis == 'y':
        axes = (0, 2)  # Rotate along the y-axis
    elif axis == 'z':
        axes = (0, 1)  # Rotate along the z-axis
    else:
        raise ValueError("Invalid rotation axis. Please choose 'x', 'y', or 'z'.")
    
#     angle = np.radians(angle)

    # Perform the rotation
    rotated_array = rotate(image_array, angle, axes=axes, reshape=False)

    # Create a new image from the rotated array
    rotated_image = Image.fromarray(rotated_array)

    # Return the rotated image
    return rotated_image

# Example usage
image_path = 'fox.jpg'  # Replace with the path to your image file
angle = 45  # Replace with the desired rotation angle (in degrees)
axis = 'z'  # Replace with the desired rotation axis ('x', 'y', or 'z')

rotated_image = rotate_image(image_path, angle, axis)
rotated_image.show()  # Display the rotated image
im1=rotated_image.save("geeks.jpg")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def show_image_3d(image_path):
    # Load the image
    image = Image.open(image_path)

    # Convert the image to a grayscale numpy array
    image_array = np.array(image.convert('L'))

    # Get the dimensions of the image
    height, width = image_array.shape

    # Create x and y coordinate arrays
    x = np.arange(width)
    y = np.arange(height)

    # Create a meshgrid from x and y
    X, Y = np.meshgrid(x, y)

    # Flatten the image array
    Z = image_array.flatten()

    # Create a 3D plot
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Plot the image as a surface
    ax.plot_surface(X, Y, Z.reshape((height, width)), cmap='gray')

    # Set labels and title
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Intensity')
    ax.set_title('Image in 3D')

    # Show the plot
    plt.show()

# Example usage
image_path = 'path_to_your_image.png'  # Replace with the path to your image file

show_image_3d(image_path)
```
