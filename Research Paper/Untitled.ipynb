```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def calculate_saliency_map(image):
    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Calculate the gradient magnitude of the grayscale image
    gradient_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
    gradient_y = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)
    gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)
    
    # Normalize the gradient magnitude to [0, 1]
    gradient_magnitude = cv2.normalize(gradient_magnitude, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)
    
    # Return the saliency map
    return gradient_magnitude

# Load the image
image_path = "text.png"
image = cv2.imread(image_path)

# Calculate the saliency map
saliency_map = calculate_saliency_map(image)

# Plot the saliency map
plt.imshow(saliency_map)
plt.colorbar()
plt.title('Saliency Map')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pytesseract

def calculate_saliency_map(image):
    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Calculate the gradient magnitude of the grayscale image
    gradient_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
    gradient_y = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)
    gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)
    
    # Normalize the gradient magnitude to [0, 1]
    gradient_magnitude = cv2.normalize(gradient_magnitude, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)
    
    # Return the saliency map
    return gradient_magnitude

def detect_text(image_path):
    # Load the image
    image = cv2.imread(image_path)
    
    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Perform text detection using pytesseract
    text = pytesseract.image_to_string(gray)
    
    return text

# Provide the path to the image containing multiple texts
image_path = "text.png"

# Load the image
image = cv2.imread(image_path)

# Calculate the saliency map
saliency_map = calculate_saliency_map(image)

# Perform text detection
text = detect_text(image_path)

# Display the saliency map
plt.imshow(saliency_map, cmap='hot')
plt.colorbar()
plt.title('Saliency Map')
plt.show()

# Print the detected text
print("Detected Text:")
print(text)
```
